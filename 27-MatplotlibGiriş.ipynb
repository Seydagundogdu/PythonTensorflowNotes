```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
%matplotlib inline
```

```python
yasList = [10,20,30,30,30,40,50,60,70,75]
```

```python
kiloList = [20,60,80,85,86,87,70,90,95,90]
```

```python
numpyYasList = np.array(yasList)
numpyKiloList = np.array(kiloList)
```

```python
numpyYasList
```

```python
numpyKiloList
```

```python
## kilonun yaşa göre değişimi. yaş = x, kilo = y
```

```python
plt.plot(numpyYasList,numpyKiloList,"b")
plt.xlabel ("Yas")
plt.ylabel("Kilo")
plt.title("Kilonun Yaşa Göre Değişim Grafiği")
```

```python
numpyDizisi1 = np.linspace(0,10,20)
```

```python
numpyDizisi1
```

```python
numpyDizisi2 = numpyDizisi1 ** 3
```

```python
numpyDizisi2
```

```python
plt.plot(numpyDizisi1,numpyDizisi2,"g--")
```

```python
## aynı anda birden çok grafik çizdirme
```

```python
plt.subplot(1,2,1) ## 1 sıra,2 kolon, 1. grafik
plt.plot(numpyDizisi1,numpyDizisi2,"r*-")
plt.subplot(1,2,2) ## 1 sıra,2 kolon, 2. grafik
plt.plot(numpyDizisi2,numpyDizisi1,"g--")
```

```python
benimFigure = plt.figure()
figureAxes = benimFigure.add_axes([0.1,0.1,0.9,0.9])
figureAxes.plot(numpyDizisi1,numpyDizisi2,"g")
figureAxes.set_xlabel("X Ekseni Veri İsmi")
figureAxes.set_ylabel("Y EkseniVeri İsmi")
figureAxes.set_title("Grafik Başlığı")
```

```python
## İç İçe Grafik
```

```python
figure2 =plt.figure()

eksen1=figure2.add_axes([0.1,0.1,0.9,0.9])
eksen2=figure2.add_axes([0.2,0.5,0.3,0.3])

eksen1.plot(numpyDizisi1, numpyDizisi2, "g")
eksen1.set_xlabel("X Ekseni")
eksen1.set_label("Y Ekseni")
eksen1.set_title("Ana Grafik Başlığı")

eksen2.plot(numpyDizisi2, numpyDizisi1, "b")
eksen2.set_xlabel("X Ekseni")
eksen2.set_label("Y Ekseni")
eksen2.set_title("Küçük Grafik Başlığı")
```

```python
## subplots'la iç içe grafik oluşturma
```

```python
plt.figure()
```

```python
plt.subplots()
```

```python
(benimFigure, benimEksenler)= plt.subplots(nrows=1, ncols=2)
for eksen in benimEksenler:
    eksen.plot(numpyDizisi1, numpyDizisi2, "b")
    eksen.set_xlabel("x ekseni")
    eksen.set_ylabel("y ekseni")
    
plt.tight_layout()
```

### İleri Görselleştirmeler


```python
yeniFigur = plt.figure(dpi=100)
## dpi = grafiğin çözünürlüğünü belrtir
yeniEksen = yeniFigur.add_axes([0.1,0.1,0.9,0.9])

yeniEksen.plot(numpyDizisi1, numpyDizisi1 ** 2,label = "numpyDizisi^2")
yeniEksen.plot(numpyDizisi1, numpyDizisi1 ** 3, label = "numpyDizisi^3")

yeniEksen.legend(loc = 2) ## çizgiler hakkında bilginin (legend) konumu
```
